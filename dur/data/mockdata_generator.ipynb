```python
from datetime import datetime
from drawdata import ScatterWidget
from sqlalchemy import create_engine
```

```python
widget = ScatterWidget(height=200, width=200)
widget
```

```python
df = widget.data_as_pandas[["x", "y", "label"]]
df["timestamp"] = datetime.now()
df.to_csv("point_samples.csv", index=False)
df.describe()
```

```python
# Configuration
server = 'localhost'
port = '1972' # Standard InterSystems superserver port
namespace = 'USER'
username = 'SuperUser'
password = 'demo'

engine = create_engine(f"iris://{username}:{password}@{server}:{port}/{namespace}")

df.to_sql("PointSamples", engine, schema="MLpipeline", if_exists='append', method='multi') # Kepps data type in DB
```
